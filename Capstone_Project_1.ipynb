```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]  # Get table from Wikipedia
```

```python
# First Level Preprocessing
df=df.drop(0)
df.columns=['PostalCode','Borough','Neighborhood']
df.drop(df[df.Borough=="Not assigned"].index,inplace=True)
df['Neighborhood']=df['Neighborhood'].fillna(df.Borough)
```

```python
#Second Level Preprocessing

new_df=df.groupby('PostalCode')['Borough','Neighborhood'].agg([('Neighborhood', ', '.join)]).reset_index()
new_df.columns=['PostalCode','Borough','Neighborhood']
new_df['Neighborhood']=new_df['Neighborhood'].replace('/',',',regex=True)
```

Work 3: Explore and cluster the neighborhoods in Toronto

```python
#only boroughs that contain the word Toronto
t_df=new_df[new_df['Borough'].str.contains("Toronto")]
```

```python
t_df
```

```python
from sklearn.cluster import KMeans
```

```python
# run k-means clustering
kmeans = KMeans(n_clusters=5, random_state=0).fit(t_df[['Latitude','Longitude']])
```

```python
t_df.insert(0,'Labels',kmeans.labels_)
```
